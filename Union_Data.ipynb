## Importamos las librerias

```python
import pandas as pd
```

## Cargamos los datos

```python
plataformas = pd.read_csv(filepath_or_buffer="Data_util/Plataformas.csv")
resenas = pd.read_csv(filepath_or_buffer="Data_util/Ratings.csv")
```

## Unimos los datos en un solo DataFrame

```python
df = plataformas.merge(resenas,left_on='id', right_on='movieId', how="outer")
```

### Visualizamos

```python
df.head()
```

```python
df.info()
```

## Elimino el campo movieId ya que se encuentra repetido en nuestro DataFrame

```python
df.drop("movieId", axis=1, inplace=True)
```

```python
df.head(1)
```

## Creo un DataFrame con el promedio de las resenas para cada pelicula

```python
score_prom = df.groupby(['id'], as_index=False).mean("score")
score_prom.drop(["release_year","duration_int", "userId"], axis=1, inplace=True)
score_prom["score"] = round(score_prom["score"],2)
score_prom.head()
```

```python
union_plataformas = plataformas.merge(score_prom,left_on='id', right_on='id', how="left")
```

```python
union_plataformas
```

```python
union_plataformas.info()
```

## Lo exporto a un archivo csv

```python
union_plataformas.to_csv("Data_util/Plataformas_score.csv", index=False)
```
